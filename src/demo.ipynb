```python
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
import analysis
import importers
from core import Core
```

```python
files: list = glob("../01UW-A/IM*")
core = importers.dicom(files=files)

slice = core.slice(axis=2, loc=16)
slice = slice[:, 775:875]

plt.figure(figsize=(10,6))
plt.imshow(slice)
plt.show()
```

```python
bt = analysis.Analysis.brightness_trace(slice=slice)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, gridspec_kw={'width_ratios': [0.5, 1.5, 1.5]}, sharey=True)

fig.set_figheight(10)
ax1.imshow(slice, cmap = 'viridis')
ax2.plot(bt[0], range(len(bt[0])))
ax3.plot(bt[1], range(len(bt[1])))

plt.show()
```
